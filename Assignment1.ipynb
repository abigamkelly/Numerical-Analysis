```python
import math
from sympy import *

def main():
    question1()
    question2()
    
    return

def question2():
    # Question 2
    # part a: Newton's Method
    print("\nQUESTION 2A:")
    p0 = 1
    tol = 1E-6
    i = 1

    while abs(math.sqrt(2) - p0) > tol:
        p = p0 - f(p0)/df(p0)
        p0 = p
        i+=1

    print("Number of iterations for Newton's Method: ", i)

    # part b: Bisection Method
    # 4 or 5 iterations???
    print("QUESTION 2B:")
    a = 1
    b = 2
    p = (a + b) / 2
    count = 1

    while abs((math.sqrt(2) - p)) > tol:
        if(f(p) == 0):
            break
        elif f(p) * f(a) > 0:
            a = p
        else:
            b = p
        
        count += 1
        p = (a + b) / 2

    print("Number of iterations for Bisection Method:", count)
    
    return
    

def question1():
    # Question 1
    # part a
    print("QUESTION 1A:")
    a = 1
    
    t = symbols('t')
    expr = (-1 / 3) * a * t * math.e**(-t / 3) + a * math.e**(-t / 3)
    sol = solve(expr)
    t_val = int(sol[0])
    print("Time of maximum:", t_val, "hours")
    
    A = symbols('A')
    expr = A * t_val * math.e**(-t_val / 3)
    sol = solve(Eq(expr, 1))
    a_val = sol[0]
    print("Amount to be injected:", a_val)

    # part b
    print("QUESTION 1B:")
    p0 = 11     #guess for p0 from graph
    tol = 1E-6

    while abs(11.077904 - p0) > tol:
        p = p0 - c(p0)/dc(p0)
        p0 = p

    print("Number of minutes: %i" %(p * 60))

    # part c
    print("QUESTION 1C:")
    t1 = p
    p0 = 21     #guess for p0 from graph
    tol = 1E-3

    while abs(21.23309 - p0) > tol:
        p = p0 - c3(p0, t1)/dc3(p0, t1)
        p0 = p

    print("Number of hours: %i" %p)
    print("Number of minutes: %i" %((p - 21) * 60 + 1))
    
    return

    

def f(x):
    return x**2 - 2

def df(x):
    return 2 * x

def c(t):
    return t - ((0.75/math.e) * math.e**(t/3))

def dc(t):
    return 1 - ((0.75/(math.e * 3)) * math.e**(t/3))

def c3(t, p):
    return (math.e/3 * t * math.e**(-t/3) + 0.75 * math.e/3 * (t - p) * math.e**((p-t)/3)) - 0.25

def dc3(t, p):
    return math.e / 4 * math.e**((p-t)/3) - math.e/12 * t * math.e**((p-t)/3) + math.e/12 * p *math.e**((p-t)/3)


main()
```
