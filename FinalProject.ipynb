```python
#h = 0.05
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# define the ODE function
def my_ode(t, y):
    x, y = y
    dxdt = x*(4 - 0.0003*x - 0.0004*y)
    dydt = y*(2 - 0.0002*x - 0.0001*y)
    return [dxdt, dydt]

# define the initial conditions and number of intervals
y0 = [10000, 10000]
num_intervals = 80

# create an array of evenly spaced time points
t_span = np.linspace(0, 4, num_intervals)

# solve the ODEs using the fourth-order Runge-Kutta method
sol = solve_ivp(my_ode, [t_span[0], t_span[-1]], y0, t_eval=t_span, method='RK45')

sol_table = np.column_stack((t_span, sol.y[0], sol.y[1])) 

# Print the solution table
print("Time    Species 1 Population  Species 2 Population")
for row in sol_table:
    print("{:.2f}    {:.2f}              {:.2f}".format(row[0], row[1], row[2]))

# print the solutions
print("Solution for x1(t):")
print(sol.y[0][-1])
print("Solution for x2(t):")
print(sol.y[1][-1])

# plot the results
plt.plot(sol.t, sol.y[0], label='Species 1')
plt.plot(sol.t, sol.y[1], label='Species 2')
plt.xlabel('Time')
plt.ylabel('Number of Species Alive')
plt.legend()
plt.show()
```
