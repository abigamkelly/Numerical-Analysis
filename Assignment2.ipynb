```python
#Natural Cubic Spline for figure 3.13 on page 156

from scipy.interpolate import CubicSpline
from scipy.interpolate import lagrange
import numpy as np
import matplotlib.pyplot as plt

#Arrays to hold the data
x = np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
y = np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25], dtype=float)

#Plot the data and draw the line
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0.9, 13.3, 100)
y_new = f(x_new)
plt.figure(figsize = (14, 8))
plt.plot(x_new, y_new, 'r')
plt.scatter(x, y)
plt.ylim(-4, 4)
plt.xlim(0, 14)
plt.title('Goose Natural Cubic Spline Interpolation')
plt.show()
```

```python
#Lagrange for figure 3.14 on page 157

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

#Lagrange function
def lagrange_poly(n, x, y, x_new):
#set interpolated value initially to zero
    p = 0
    #implementing Lagrange Interpolation
    for i in range(n):
        L = 1
        for j in range(n):
            if i != j:
                L = (x_new - x[j])/(x[i]-x[j]) * L
        p = p + L*y[i]
    return p

#Arrays to hold data
x = np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
y = np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25], dtype=float)#corresponding temperatures
n = x.shape[0]

#Plot the points and draw the line
y_new = lagrange_poly(n, x, y, x_new)
fig = plt.figure(figsize = (14,8))
plt.plot(x_new, y_new,'b', x, y,'bo')
plt.title('Goose Lagrange Interpolation')
```

```python
#Natural Cubic Spline for dog on page 161

import numpy as np
import matplotlib.pyplot as plt

#Arrays to hold data
x1 = [1, 2, 5, 6, 7, 8, 10, 13, 17]
y1 = [3, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5]
x2 = [17, 20, 23, 24, 25, 27, 27.7]
y2 = [4.5, 7, 6.1, 5.6, 5.8, 5.2, 4.1]
x3 = [27.7, 28, 29, 30]
y3 = [4.1, 4.3, 4.1, 3]

#Plot points and draw line
f1 = CubicSpline(x1, y1, bc_type='natural')
f2 = CubicSpline(x2, y2, bc_type='natural')
f3 = CubicSpline(x3, y3, bc_type='natural')
fig = plt.figure(figsize = (14,8))
x_new1 = np.linspace(1, 17, 100)
y_new1 = f1(x_new1)
plt.plot(x_new1, f1(x_new1), 'b', x1, y1, 'ro')
x_new2 = np.linspace(17, 27.7, 100)
y_new2 = f2(x_new2)
plt.plot(x_new2, f2(x_new2), 'b', x2, y2, 'ro')
x_new3 = np.linspace(27.7, 30, 100)
y_new3 = f3(x_new3)
plt.plot(x_new3, f3(x_new3), 'b', x3, y3, 'ro')

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
plt.ylim(0, 8)
plt.xlim(0, 30)

plt.title('Natural Cubic Spline Interpolation')
plt.show()
```
